```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import math as math
```

```python
%matplotlib inline
```

```python
pokemon_df = pd.read_csv('./DataSets/Pokemon.csv')
pd.read_csv('./DataSets/Pokemon.csv')
```

```python
# Group Pokémon by generation and calculate the cumulative count
generation_counts = pokemon_df.groupby('Generation')['#'].max().cumsum()
```

```python
# Create a bar plot
plt.figure(figsize=(10, 5))
plt.bar(generation_counts.index, generation_counts.values, color='skyblue', edgecolor='black')
plt.xlabel('Generation')
plt.title('Cumulative Pokémon Count by Generation')
plt.xticks(range(1, len(generation_counts) + 1), rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)


for x, y in zip(generation_counts.index, generation_counts.values):
    plt.annotate(f'{y}', (x, y), textcoords="offset points", xytext=(0, 5), ha='center')


plt.ylim(bottom=0)

plt.show()
```
