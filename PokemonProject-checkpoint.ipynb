```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
pokemon_df = pd.read_csv('./DataSets/Pokemon.csv')
pd.read_csv('./DataSets/Pokemon.csv')
```

```python
# Group Pokemon by generation and count the number of Pokemon in each generation
generation_counts = pokemon_df.groupby('Generation')['#'].nunique()
```

```python
# Create a bar plot
plt.figure(figsize=(10, 5))
plt.bar(generation_counts.index, generation_counts.values, color='skyblue', edgecolor='black')
plt.xlabel('Generation')
plt.ylabel('Number of New PokÃ©mon')
plt.title('Number of New PokÃ©mon Added by Generation')
plt.xticks(range(1, len(generation_counts) + 1), rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on top of each bar
for x, y in zip(generation_counts.index, generation_counts.values):
    plt.annotate(f'{y}', (x, y), textcoords="offset points", xytext=(0, 5), ha='center')

plt.ylim(bottom=0)
plt.show()
```

```python
# Separate legendary and non-legendary Pokemon
legendary = pokemon_df[pokemon_df['Legendary'] == True]
non_legendary = pokemon_df[pokemon_df['Legendary'] == False]
```

```python
# Calculate summary statistics for each group
legendary_stats = legendary[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].describe()
non_legendary_stats = non_legendary[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].describe()
```

```python
# Display summary statistics
print('Legendary Pokemon Stats:')
print(legendary_stats)
print('\nNon-Legendary Pokemon Stats:')
print(non_legendary_stats)
```

```python
# Create a new column with descriptive labels
pokemon_df['LegendaryLabel'] = pokemon_df['Legendary'].apply(lambda x: 'Legendary' if x else 'Non-Legendary')

# Visualize distributions using box plots
plt.figure(figsize=(12, 8))
plt.subplot(2, 3, 1)
sns.boxplot(data=pokemon_df, x='LegendaryLabel', y='HP')
plt.subplot(2, 3, 2)
sns.boxplot(data=pokemon_df, x='LegendaryLabel', y='Attack')
plt.subplot(2, 3, 3)
sns.boxplot(data=pokemon_df, x='LegendaryLabel', y='Defense')
plt.subplot(2, 3, 4)
sns.boxplot(data=pokemon_df, x='LegendaryLabel', y='Sp. Atk')
plt.subplot(2, 3, 5)
sns.boxplot(data=pokemon_df, x='LegendaryLabel', y='Sp. Def')
plt.subplot(2, 3, 6)
sns.boxplot(data=pokemon_df, x='LegendaryLabel', y='Speed')
plt.tight_layout()
plt.show()
```

```python
# Count the number of Pokemon with each type combination
type_combinations = pokemon_df[['Type 1', 'Type 2']].dropna().apply(tuple, axis=1).value_counts()
sorted_combinations = type_combinations.sort_values(ascending=False)
# Select the top 20 most common type combinations
top_combinations = sorted_combinations.head(10)
```

```python
# Create a horizontal bar plot to visualize the most common type combinations
plt.figure(figsize=(10, 6))
top_combinations.plot(kind='barh')
plt.title('Most Common Dual-Type Pokemon Combinations')
plt.xlabel('Count')
plt.tight_layout()
plt.show()
```
